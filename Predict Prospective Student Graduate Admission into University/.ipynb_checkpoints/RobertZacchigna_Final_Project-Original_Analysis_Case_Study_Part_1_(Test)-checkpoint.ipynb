## Original Case Study Idea Using U.S. College Data

For my original case analysis I decided to take a look at the graduation rates of U.S. based public and private universities to see the distribution of universities (public and private) that have a graduation of at least 50%.

Determining what has the biggest impact on student graduation rates could help universities decide how to balance the costs for students, rebalance budgets for different departments and what faculty they need or don't need.

I found my dataset on Kaggle (https://www.kaggle.com/yashgpt/us-college-data) that contains data for 777 different United States based universities, with 18 different variables. All records in the dataset are complete, there is no missing data that needs to be accounted for in my analysis.

## Step 1: Load data from College_Data.csv

```python
collegeData = pd.read_csv('College_Data.csv')
```

### Create GradRate50Perc column

```python
rate = []
for val in collegeData['Grad.Rate']:
    if val >= 50:
        rate += ['Yes']
    else:
        rate += ['No']

collegeData['GradRate50Perc'] = rate
```

## Step 6: Histograms of my data - Room.Board, Books, Personal and Expend

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.8, wspace=0.5)
fig.set_size_inches(13.5, 10)
sb.set(font_scale = 1.15)

feats = ['Room.Board', 'Books', 'Personal', 'Expend']
histLabels = ['Room and Board Costs (U.S. Dollars)', 'Estimated Book Costs (U.S. Dollars)', 
              'Estimated Personal Spending (U.S. Dollars)', 'Instructional Expenditure (U.S. Dollars)']

i = 1
for var in feats:
    fig.add_subplot(2, 2, i)
    sb.distplot(pd.Series(collegeData[var].dropna(), name=histLabels[i-1]),
                fit=norm, kde=False).set_title(var + " Histogram")
    plt.ylabel('Count')
    
    i += 1

fig.tight_layout()
```

## Step 7: Make some bar charts for variables with only a few options.

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.8, wspace=0.5)
fig.set_size_inches(13.5, 10)
sb.set(font_scale = 1.5)

fig.add_subplot(1, 2, 1)

X_Private = collegeData.groupby('Private').size().reset_index(name='Counts')['Private']
Y_Private = collegeData.groupby('Private').size().reset_index(name='Counts')['Counts']

splot = sb.barplot(X_Private, Y_Private)

for p in splot.patches:
    splot.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')

splot.set_title('Universities')
splot.set_xticklabels(['Public', 'Private'])
plt.xlabel('')

fig.add_subplot(1, 2, 2)

X_grad50 = collegeData.groupby('GradRate50Perc').size().reset_index(name='Counts')['GradRate50Perc']
Y_grad50 = collegeData.groupby('GradRate50Perc').size().reset_index(name='Counts')['Counts']

splot = sb.barplot(X_grad50, Y_grad50)

for p in splot.patches:
    splot.annotate(format(p.get_height(), '.0f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 9), 
                   textcoords = 'offset points')

splot.set_title('Graduation Rate of at Least 50%')
plt.xlabel('')
```

## Step 8: Check data correlation with Pearson Ranking Chart

```python
plt.rcParams['figure.figsize'] = (16, 10)

sb.set(font_scale = 1.5)
sb.set_style(style='white')

columns = ['Room.Board', 'Books', 'Personal', 'Expend']
columnArrays = np.matrix(collegeData[columns].dropna())

corrData = Rank2D(features=columns, algorithm='pearson')

corrData.fit(columnArrays)
corrData.transform(columnArrays)
corrData.poof()
```

## Step 9: Use Parallel Coordinates visualization to compare the distributions of numerical variables between universities with graduation rates of at least 50%

```python
plt.rcParams['figure.figsize'] = (16, 10)

sb.set(font_scale = 1.5)
sb.set_style(style='white')

set_palette('sns_bright')

classes = ['Grad Rate < 50%', 'Grad Rate >= 50%']
columns = ['Room.Board', 'Books', 'Personal', 'Expend']

collegeDataNorm = collegeData.copy()

for col in columns:
    collegeDataNorm[col] = ((collegeData[col] - collegeData[col].min(skipna=True)) / 
                            (collegeData[col].max(skipna=True) - collegeData[col].min(skipna=True)))

x = collegeDataNorm[columns].dropna()
y = collegeData['GradRate50Perc']

parrCorrData = ParallelCoordinates(classes=classes, features=columns)

parrCorrData.fit_transform(x, y)
parrCorrData.poof()
```

## Step 10: Using Stacked Bar Charts to compare public and private universities that had a graduation rate of at least 50% based on the other variables.

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.8, wspace=0.5)
fig.set_size_inches(14, 12)

i = 1
for var in ['Grad.Rate', 'PhD', 'perc.alumni']:
    fig.add_subplot(3, 1, i)
    
    public = collegeData[collegeData['GradRate50Perc']=='Yes'][collegeData['Private']=='No'][var].value_counts()
    private = collegeData[collegeData['GradRate50Perc']=='Yes'][collegeData['Private']=='Yes'][var].value_counts()
    private = private.reindex(index = public.index)

    topPlot = plt.bar(public.index, public.values)
    bottomPlot = plt.bar(private.index, private.values, bottom=public.values)

    plt.ylabel('Counts', fontsize=15)
    plt.title(var, fontsize=15)
    plt.tick_params(axis='both', labelsize=15)
    plt.legend((topPlot, bottomPlot), ('Public', 'Private'), fontsize = 15)
    
    i += 1

fig.tight_layout()
```
